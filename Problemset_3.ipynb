# Q 1:

```python
#Bring in required libraries 
import pandas as pd

# Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|', index_col='user_id')

#Find out the average age of each profession
mean_age_by_occupation = users.groupby('occupation')['age'].mean()

#Discover the Male ratio per occupation and sort it
male_ratio_by_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count()).sort_values(ascending=False)

#For each occupation, calculate the minimum and maximum ages
min_age_by_occupation = users.groupby('occupation')['age'].min()
max_age_by_occupation = users.groupby('occupation')['age'].max()

#For each combination of occupation and sex, calculate the mean age
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()

#For each occupation, present the percentage of women and men
gender_percentage_by_occupation = users.groupby(['occupation', 'gender']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

#As needed, print the results or utilize them
print("Mean Age per Occupation:")
print(mean_age_by_occupation)
print("\nMale Ratio per Occupation (sorted):")
print(male_ratio_by_occupation)
print("\nMinimum Age by Occupation:")
print(min_age_by_occupation)
print("\nMaximum Age by Occupation:")
print(max_age_by_occupation)
print("\nMean Age per Occupation and Sex:")
print(mean_age_by_occupation_sex)
print("\nPercentage of Women and Men per Occupation:")
print(gender_percentage_by_occupation)
```

```python
#Import necessary libraries
import pandas as pd

#Import the dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

#Select only the Goal column
goal_column = euro12['Goals']

#What number of teams competed in Euro2012?
num_teams_participated = euro12['Team'].nunique()

#Number of columns in the dataset
num_columns = len(euro12.columns)

#See just the Team, Red Cards, and Yellow Cards columns
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

#Sort the teams by Red Cards, then by Yellow Cards
discipline_sorted = discipline.sort_values(by=['Red Cards','Yellow Cards'])

#Determine how many Yellow Cards each team averaged
mean_yellow_cards = euro12['Yellow Cards'].mean()

#Filter teams that scored more than 6 goals
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]

#Choose the teams whose names begin with 'G'
teams_starting_with_G = euro12[euro12['Team'].str.startswith('G')]

#Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]

#Select all columns except the last 3
all_columns_except_last_3 = euro12.iloc[:, :-3]

#Shooting Accuracy from Russia, England, and Italy are presented
shooting_accuracy_selected_teams = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]

print("Goal Column:")
print(goal_column)
print("\nNumber of Teams Participated in Euro2012:")
print(num_teams_participated)
print("\nNumber of Columns in the Dataset:")
print(num_columns)
print("\nDiscipline DataFrame (Team, Yellow Cards, Red Cards):")
print(discipline)
print("\nDiscipline DataFrame Sorted by Red Cards, then Yellow Cards:")
print(discipline_sorted)
print("\nMean Yellow Cards Given per Team:")
print(mean_yellow_cards)
print("\nTeams that Scored More than 6 Goals:")
print(teams_more_than_6_goals)
print("\nTeams Starting with 'G':")
print(teams_starting_with_G)
print("\nFirst 7 Columns:")
print(first_7_columns)
print("\nAll Columns Except the Last 3:")
print(all_columns_except_last_3)
print("\nShooting Accuracy for England, Italy, and Russia:")
print(shooting_accuracy_selected_teams)
```

```python
#Import libraries
import pandas as pd

#A suitable datetime index should be used to replace the first three columns. Assign it to a variable named data
with open('wind.txt') as file:
    data =pd.read_csv(file, delim_whitespace=True, na_values='NaN')
    data['Yr_Mo_Dy']=pd.to_datetime(data[['Yr','Mo','Dy']].astype(str).agg('-'.join,axis=1),errors='coerce')
    data =data.set_index('Yr_Mo_Dy')

#For numbers higher than 2000, set the year
data.index=data.index.to_series().apply(lambda x: x.replace(year=x.year - 100) if x.year > 2000 else x)

#Use the datetime64[ns] type to set the correct dates as the index
data.index =pd.to_datetime(data.index, errors='coerce')

#Count missing values for each location
missing_values_per_location = data.isnull().sum()

#Count non-missing values in total
non_missing_values_total =data.count().sum()

#Determine the dataset's average windspeeds
mean_windspeed_entire_dataset = data.mean().mean()

#Create a DataFrame called loc_stats and calculate min, max, mean, and std at each location
loc_stats = data.agg(['min','max','mean','std'])

#Make a DataFrame named day_stats, then compute the mean, standard deviation, min, and max for each day
day_stats = data.resample('D').agg(['min','max','mean','std'])

#Find the average windspeed in January for each location
average_windspeed_january =data[data.index.month==1].groupby(data[data.index.month==1].index.year).mean()

#Reduce the record's sampling frequency to once a year for every location
yearly_data = data.resample('Y').mean()

#Downsample the record to a monthly frequency for each location
monthly_data = data.resample('M').mean()

#Reduce the record's frequency to once a week for every site
weekly_data = data.resample('W').mean()

#Determine the first 52 weeks' min, max, mean, and standard deviation
weekly_stats = data.resample('W-Mon').agg(['min', 'max', 'mean', 'std']).iloc[:52]

print("Missing values per location:\n", missing_values_per_location)
print("\nNon-missing values in total:", non_missing_values_total)
print("\nMean windspeed for the entire dataset:", mean_windspeed_entire_dataset)
print("\nLocation statistics:\n", loc_stats)
print("\nDay statistics:\n", day_stats)
print("\nAverage windspeed in January for each location:\n", average_windspeed_january)
print("\nYearly data:\n", yearly_data)
print("\nMonthly data:\n", monthly_data)
print("\nWeekly data:\n", weekly_data)
print("\nWeekly statistics for the first 52 weeks:\n", weekly_stats)
```

```python
#Import necessary libraries
import pandas as pd

#Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url,sep='\t')

#Assign it to a variable called chipo
#View the initial ten entries
print(chipo.head(10))

#How many observations are there in the dataset
num_observations=chipo.shape[0]

#How many columns does the dataset contain
num_columns=chipo.shape[1]

#Print the name of all the columns
columns_names=chipo.columns.tolist()
#How is the dataset indexed?
index_type =chipo.index
#Which was the most-ordered item?
most_ordered_item =chipo['item_name'].mode().iloc[0]

#How many things were ordered for the most-requested item
num_most_ordered_item =chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
#What was the most ordered item in the choice_description column
most_ordered_choice= chipo['choice_description'].mode().iloc[0]

#How many items were ordered in total?
total_items_ordered= chipo['quantity'].sum()

#Make the item's pricing a floating value
chipo['item_price'] =chipo['item_price'].apply(lambda x: float(x[1:]))

#Check the item price type
item_price_type_before =chipo['item_price'].dtype
#Make a lambda function and modify the item price type
chipo['item_price'] =chipo['item_price'].apply(lambda x: float(x))

#Check the item price type
item_price_type_after =chipo['item_price'].dtype
#How much was the revenue for the period in the dataset?
revenue = (chipo['quantity']*chipo['item_price']).sum()

#During that time, how many orders were placed
num_orders = chipo['order_id'].nunique()
#What is the average revenue amount per order
avg_revenue_per_order = revenue/num_orders

#Get variety of products are sold
num_unique_items =chipo['item_name'].nunique()

print("Number of Observations:", num_observations)
print("Number of Columns:", num_columns)
print("Columns Names:", columns_names)
print("Index Type:", index_type)
print("Most-Ordered Item:", most_ordered_item)
print("Quantity of Most-Ordered Item:", num_most_ordered_item)
print("Most Ordered Choice Description:", most_ordered_choice)
print("Total Items Ordered:", total_items_ordered)
print("Item Price Type Before Conversion:", item_price_type_before)
print("Item Price Type After Conversion:", item_price_type_after)
print("Revenue for the Period:", revenue)
print("Number of orders:", num_orders)
print("Average Revenue amount per Order:", avg_revenue_per_order)
print("Number of Different Items Sold:", num_unique_items)
```

```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#Import the dataset
file_path="us-marriages-divorces-1867-2014.csv"
data=pd.read_csv(file_path)

#Create a line plot
plt.figure(figsize=(10, 6))

#Chart the amount of weddings per 1,000 people
plt.plot(data['Year'],data['Marriages_per_1000'], label='Marriages per Capita',color='blue')

#Plot the divorce rate by population
plt.plot(data['Year'], data['Divorces_per_1000'],label='Divorces per Capita',color='skyblue')

#Label axes and add legend
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

#Show the plot
plt.show()
```

```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#Import the dataset
file_path ="us-marriages-divorces-1867-2014.csv"
data= pd.read_csv(file_path)

#Sort and filter information for 1900, 1950, and 2000
selected_years = [1900, 1950, 2000]
selected_data =data[data['Year'].isin(selected_years)]

#Create a vertical bar chart
plt.figure(figsize=(8, 4))

#Plot the number of marriages per capita
plt.bar(selected_data['Year']-3,selected_data['Marriages_per_1000'],width=5,label='Marriages per Capita',color='blue')

#Plot the number of divorces per capita
plt.bar(selected_data['Year']+3,selected_data['Divorces_per_1000'],width=5,label='Divorces per Capita', color='skyblue')

#Legend added, axes labeled
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks(selected_data['Year'])
plt.legend()

#Show the plot
plt.show()
```

```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#Import the dataset
file_path= "actor_kill_counts.csv"
data =pd.read_csv(file_path)

#Arrange the actors in order of number of kills
data.sort_values(by='Count',ascending=True,inplace=True)

#Make a bar chart that is horizontal
plt.figure(figsize=(10, 6))
plt.barh(data['Actor'],data['Count'],color='skyblue')

#Label axes and add title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')

#Add labels to each bar
for index, value in enumerate(data['Count']):
    plt.text(value,index,str(value))

#Show the plot
plt.show()
```

```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#Import the dataset
file_path ="roman-emperor-reigns.csv"
data =pd.read_csv(file_path)

#Calculate the number of emperors who were assassinated
assassinated_emperors =data[data['Cause_of_Death']=='Assassinated']
num_assassinated =len(assassinated_emperors)

#Determine the total number of emperors who escaped assassination
num_not_assassinated = len(data)-num_assassinated

#Create a pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [num_assassinated, num_not_assassinated]
colors = ['blue', 'lightblue']

plt.figure(figsize=(4,4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',colors=colors,startangle=90)

#In the middle of the pie chart, draw a circle
centre_circle =plt.Circle((0, 0),0.80,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

#A pie with an equal aspect ratio illustrated as a circle
plt.axis('equal')
plt.title('Fraction of Roman Emperors Assassinated')
plt.show()
```
